# Matplotlib Intro

![matplotlib logo](../images/matplotlib_logo.png)

The oldest plotting library in Python - created as a way to get Matlab programmers enticed over to Python, we owe a huge debt of gratitude to this venerable old library. Matplotlib can do pretty much anything, it can just take many lines of code to get there.

One of the main points of confusion in matplotlib is that it has two APIs, an OOP-based API and the pyplot API. (We don't talk about pylab...) 
The pyplot API was designed to mimic Matlab and is great for simple plots, but can be a bit "magical" if you want slightly more complicated plots

Let's start by looking at the pyplot API, usually imported as `plt`

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 20)
y = x ** 2
```

```python
plt.plot(x, y);
plt.show()
```

Pretty simple - now let's do a bar chart

```python
plt.bar(x, y)
plt.show()
```

Let's try the same with the Object-Oriented approach 

**Note: I have never done this before and neither should you**

```python
import matplotlib as mpl
from matplotlib.backends.backend_agg import FigureCanvas

fig = mpl.figure.Figure()
canvas = FigureCanvas(fig)
ax = fig.add_subplot()

ax.plot(x, y)
fig.savefig('test.png')
```

Usually, we mix just a little bit, since the plt API gives you some really nice abstractions to get you started, and then keep working in the OOP approach.

Let's do it the right way this time:

```python
fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
fig, ax = plt.subplots()
ax.bar(x, y)
```

That little `subplots` part, basically sets up the tedious stuff I couldn't get to work before - creating a figure, setting the backend and renderer and returning a configured figure object and axis

Conceptually, the `figure` is the piece of paper we're drawing on, and the `ax` is the actual graph we're drawing. 

This abstraction lets us do plots like this:

```python
fig = plt.figure()
ax = fig.add_subplot(2, 2, 1)
ax1 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax3.hist(y, bins=5)
ax.bar(x, y)
ax1.plot(x, y)
```

Matplotlib is very empirical - you run through the steps you want it to draw and it draws them.
If you want a histogram, you tell it you want a histogram:

```python
fig, ax = plt.subplots()
ax.hist(y, bins=5);
```

Styling is the same, you specify the attributes you want to modify:

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
ax1.bar(x, y, color="red")
ax2.plot(x, y, linestyle='dashdot')
ax1.set_title("My lovely Barchart")
ax2.set_title("Another Easy Graph")
ax1.set_ylabel("Some count")
```

We can add annotations to make it superfancy!

```python
fig, ax = plt.subplots()
ax.plot(x, y, linestyle='--')
ax.annotate("This part looks super interesting!", 
            xy=(x[5], y[5]),
            xytext=(1, 300), 
            arrowprops=dict(arrowstyle='fancy', connectionstyle="arc3,rad=0.3"),
           fontsize=15);
```
